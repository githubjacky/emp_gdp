##  1. Simulate a time series ùë¶ùë° of length ùëõ = 100 following an ARMA(1,1) model with ùúô = 0.8 and ùúÉ = 0.4.

```python
import statsmodels
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

from rpy2 import robjects
from rpy2.robjects.packages import importr
```

### (a) Calculate and plot the theoretical autocorrelation function for this model. Plot sufficient lags until the correlations are negligible.

```python
ARMAacf = importr('stats').ARMAacf

res = robjects.r('''
    ARMAacf(ar=c(0.8), ma=c(0.4), lag.max=20)
''')
plt.plot(res)
```

### (b) Calculate and plot the sample ACF for your simulated series. How well do the values and patterns match the theoretical ACF from part (a)?

```python
phi = [1, -0.8]
theta = [1, 0.4]
```

```python
np.random.seed(12)
samples = sm.tsa.arma_generate_sample(ar=phi, ma=theta, nsample=100)
statsmodels.graphics.tsaplots.plot_acf(samples, lags=14)
print("acf: ", statsmodels.tsa.stattools.acf(samples, nlags=14))
```

### (c) Calculate and interpret the sample EACF for this series. Does the EACF help you specify the correct orders for the model?

```python
tsa = importr('TSA')
```

```python
tsa.eacf(robjects.FloatVector(samples))
```

since the result is quite confusing, it's not convincing to determine the right order

### (d) Repeat parts (b) and (c) with a new simulation using the same parameter values but sample size ùëõ = 48.

```python
np.random.seed(12)
samples = sm.tsa.arma_generate_sample(ar=phi, ma=theta, nsample=48)
statsmodels.graphics.tsaplots.plot_acf(samples, lags=12)
print("acf: ", statsmodels.tsa.stattools.acf(samples, nlags=12))
```

```python
tsa.eacf(robjects.FloatVector(samples))
```

### (e) Repeat parts (b) and (c) with a new simulation using the same parameter values but sample size ùëõ = 200.

```python
np.random.seed(12)
samples = sm.tsa.arma_generate_sample(ar=phi, ma=theta, nsample=200)
statsmodels.graphics.tsaplots.plot_acf(samples, lags=13)
print("acf: ", statsmodels.tsa.stattools.acf(samples, nlags=13))
```

## 3. The data file named robot contains a time series obtained from an industrial robot. The robot was put through a sequence of maneuvers, and the distance from a desired ending point was recorded in inches.

### (a) Display the time series plot of the data. Based on this information, do these data appear to come from a stationary or nonstationary process?

```python
import csv
with open('TSA HW06.robot.csv', newline='') as f:
    reader = csv.reader(f)
    data = [i for i in reader]
data = [i[0] for i in data][1:]
data = [float(i) for i in data]

plt.plot(data)
```

non-stationary

### (b) Calculate and plot the sample ACF and PACF for these data. Based on this additional information, do these data appear to come from a stationary or nonstationary process?

```python
statsmodels.graphics.tsaplots.plot_acf(data, lags=15)
statsmodels.graphics.tsaplots.plot_pacf(data, lags=15);
```

it seems non stationary for the existence of the auto correlation

### (c) Calculate and interpret the sample EACF.

```python
tsa.eacf(robjects.FloatVector(data))
```

it's quite possible that the order of AR is high and so does MA

### (d) Estimate the parameters of an AR(1) model and IMA(1, 1) for these data, respectively.

```python
model = statsmodels.tsa.arima.model.ARIMA(data, order=(1, 0, 0)).fit()
model.params
```

```python
ar1_ll = statsmodels.tsa.arima.model.ARIMA(data, order=(1, 0, 0)).loglike(model.params)
```

```python
model = statsmodels.tsa.arima.model.ARIMA(data, order=(0, 1, 1)).fit()
model.params
```

```python
ima_ll = statsmodels.tsa.arima.model.ARIMA(data, order=(0, 1, 1)).loglike(model.params)
```

### (e) Compare the results from parts (d) in terms of AIC and discuss the residual tests.

```python
print('AIC of AR1: ', statsmodels.tools.eval_measures.aic(ar1_ll, len(data), 2))
```

```python
print('AIC of IMA11: ', statsmodels.tools.eval_measures.aic(ima_ll, len(data), 3))
```

because the loss for IMA(1, 1) is smaller, it's bettter to choose it
