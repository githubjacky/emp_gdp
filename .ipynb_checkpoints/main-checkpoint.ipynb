```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (20,5)
```

```python
dfs = pd.read_csv('data/SAEMP_GDP.csv')
dfs.dropna(inplace=True)
```

```python
set(dfs['timeperiod'])
```

```python
dfs.columns
```

```python
geonames = list(set(dfs['geoname']))
num_geo = len(geonames)
pct_changes_emp = np.zeros(num_geo)

for i in range(num_geo):
    pct_change = dfs[dfs['geoname'] == geonames[i]]['employment'].pct_change().iloc[1:]
    pct_changes_emp[i] = np.mean(pct_change)
```

```python
idx = np.argsort(pct_changes_emp)
idx_min = idx[0]
print('minimum change: ', geonames[idx_min])
idx_max = idx[-1]
print('maximum change: ', geonames[idx_max])
```

```python
pct_changes_gdp = np.zeros(num_geo)

for i in range(num_geo):
    pct_change = dfs[dfs['geoname'] == geonames[i]]['GDP'].pct_change().iloc[1:]
    pct_changes_gdp[i] = np.mean(pct_change)
```

```python
idx = np.argsort(pct_changes_gdp)
idx_min = idx[0]
print('minimum change: ', geonames[idx_min])
idx_max = idx[-1]
print('maximum change: ', geonames[idx_max])
```

```python
gdpemp_coef = np.zeros(num_geo)
for i in range(num_geo):
    pct_change_emp = dfs[dfs['geoname'] == geonames[i]]['employment'].pct_change().iloc[1:].to_numpy()
    pct_change_gdp = dfs[dfs['geoname'] == geonames[i]]['GDP'].pct_change().iloc[1:].to_numpy()
    gdpemp_coef[i] = np.corrcoef(pct_change_emp[1:], pct_change_gdp[:-1])[0, 1]
```

```python
plt.rcParams["figure.figsize"] = (20,15)

plt.subplot(211)
plt.ylabel('correlation coefficient', fontsize=25)
plt.xlabel('employment growth rate(100%)', fontsize=25)
plt.scatter(pct_changes_emp*100, gdpemp_coef)

plt.subplot(212)
plt.ylabel('correlation coefficient', fontsize=25)
plt.xlabel('GDP growth rate(100%)', fontsize=25)
plt.scatter(pct_changes_gdp*100, gdpemp_coef)
plt.savefig('report/res/corr.png')
```

```python
ave_gdp = np.zeros(num_geo)

for i in range(num_geo):
    mean = dfs[dfs['geoname'] == geonames[i]]['GDP'].mean()
    ave_gdp[i] = mean / 1e9
```

```python
ave_emp = np.zeros(num_geo)

for i in range(num_geo):
    mean = dfs[dfs['geoname'] == geonames[i]]['employment'].mean()
    ave_emp[i] = mean / 1e6
```

```python
pd.DataFrame.from_dict({
    'geoname': geonames,
    'gdp_gr': pct_changes_gdp*100,
    'emp_gr': pct_changes_emp*100,
    'ave_gdp': ave_gdp,
    'ave_emp': ave_emp,
    'lag_corr': gdpemp_coef 
}).to_csv('data/corr.csv')
```
