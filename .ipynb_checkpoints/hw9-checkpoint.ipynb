```python
from scipy.fft import fft
from scipy.signal import stft
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

## 1. (20%) Simulate three random variables with length 1024, following standard Normal, ùë°-distribution (ùëëùëì = 10), and exponential distribution (rate = 1.6), respectively.

```python
y1 = np.random.standard_normal(size=1024)
y2 = np.random.standard_t(10, size=1024)
y3 = np.random.exponential(1.6, size=1024)
```

#### (a) Perform FFT (Fast Fourier Transform) over the three random variables and plot the amplitudes.

```python
plt.rcParams["figure.figsize"] = (20,5)

plt.subplot(311)
plt.plot(fft(y1))
plt.title('standard normal')
plt.subplot(312)
plt.plot(fft(y2))
plt.title('t-distribution(10)')
plt.subplot(313)
plt.plot(fft(y3))
plt.title('exponential(1.6)');
```

#### (b) Perform STFT (Short-Time Fourier Transform) over the three random variables and plot the timefrequency contours.

```python
plt.rcParams["figure.figsize"] = (30,30)
plt.subplot(311)
plt.contourf(stft(y1)[2])
plt.title('standard normal')
plt.subplot(312)
plt.contourf(stft(y2)[2])
plt.title('t-distribution(10)')
plt.subplot(313)
plt.contourf(stft(y3)[2])
plt.title('exponential(1.6)');
```

#### (c) What do you observe in (a) and (b)?

purely noisy?

## 2. (20%) Simulate a seasonal time series following the model SARIMA(2, 1, 0) √ó (0, 1, 1)12.

```python
from rpy2.robjects.packages import importr
from rpy2 import robjects
```

```python
sarima = importr('astsa')
```

```python
samples = robjects.r('''
    sarima.sim(ar=0.8, d=1, ma=0, sar=0, D=1, sma=0.8, S=12, n=1024)
''')
```

#### (a) Perform FFT (Fast Fourier Transform) over the time series and plot the amplitudes.

```python
plt.rcParams["figure.figsize"] = (20,5)

plt.plot(fft(samples));
```

#### (b) Perform STFT (Short-Time Fourier Transform) over the time series and plot the time-frequency contours.

```python
plt.contourf(stft(samples)[2]);
```
